```python
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt

symbol = "EURUSD=X"
start_date = "2023-03-01"
end_date = "2023-03-31"
timeframe = "5m"

data = yf.download(symbol, start=start_date, end=end_date, interval=timeframe)

highs = data['High']
lows = data['Low']
close = data['Close']

def fibonacci(highs, lows, close, factor):
    last_close = close[-1]
    first_high = highs[0]
    first_low = lows[0]
    price_range = first_high - first_low
    support_level = last_close - factor * price_range
    resistance_level = last_close + factor * price_range
    return [support_level, resistance_level]

fibo_levels = fibonacci(highs, lows, close, 1.618)

results = []
for i in range(2, len(data)):
    if data['Low'][i] < fibo_levels[0] and data['Close'][i] > fibo_levels[0]:
        # Se o preço cruzou acima do nível de Fibonacci 61.8%, abrimos uma posição de compra
        buy_price = data['Close'][i]
        stop_loss = buy_price - 10
        take_profit = buy_price + 15
        
        # Verificamos se a posição atingiu o stop loss ou take profit
        j = i + 1

        while j < len(data) and (data['High'][j] < take_profit and data['Low'][j] > stop_loss):
            j += 1
        if j < len(data) and data['High'][j] >= take_profit:
            # A posição atingiu o take profit
            results.append(take_profit - buy_price)
        else:
            # A posição atingiu o stop loss
            results.append(stop_loss - buy_price)
            
# Calculamos a rentabilidade da estratégia
total_return = np.product(1 + np.array(results)) - 1

# Calculamos o capital final
fator_crescimento = 1 + total_return
capital_final = 10 * fator_crescimento

rentabilidade_pct = round(total_return * 100, 2)
capital_final_milhoes = round(capital_final / 1000000, 2)

print("Rentabilidade:", rentabilidade_pct, "%")
print("Capital final: R$", capital_final_milhoes, "milhões")


# Plotamos um gráfico da estratégia
fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(data['Close'], label='Preço de Fechamento')
ax.axhline(y=fibo_levels[0], linestyle='--', color='gray', label='Nível de Fibonacci 61.8%')
ax.legend()
ax.set_xlabel('Data')
ax.set_ylabel('Preço')
ax.set_title('Backtest da Estratégia')
plt.show()
```

```python
import pandas as pd
import numpy as np
import yfinance as yf

symbol = "EURUSD=X"
start_date = "2023-03-01"
end_date = "2023-03-31"
timeframe = "15m"

data = yf.download(symbol, start=start_date, end=end_date, interval=timeframe)

# Calcular as médias móveis
ema12 = data['Close'].ewm(span=12, adjust=False).mean()
ema50 = data['Close'].ewm(span=50, adjust=False).mean()

results = []
for i in range(50, len(data)):
    # Verificar a posição das médias móveis
    if ema12[i] > ema50[i] and ema12[i-1] <= ema50[i-1]:
        # Se a EMA 12 cruza acima da EMA 50, abrimos uma posição de compra
        buy_price = data['Close'][i]
        stop_loss = buy_price - 10
        take_profit = buy_price + 20
        
        # Verificamos se a posição atingiu o stop loss ou take profit
        j = i + 1
        while j < len(data) and (data['High'][j] < take_profit and data['Low'][j] > stop_loss):
            j += 1
        
        if j < len(data) and data['High'][j] >= take_profit:
            # A posição atingiu o take profit
            results.append(take_profit - buy_price)
        else:
            # A posição atingiu o stop loss
            results.append(stop_loss - buy_price)

# Calculamos a rentabilidade da estratégia
total_return = np.product(1 + np.array(results)) - 1
capital_final = 1 * (1 + total_return)

print("Rentabilidade:", round(total_return * 100, 2), "%")
print("Capital final: R$", round(capital_final, 2))
```

```python
import pandas as pd
import numpy as np
import yfinance as yf
import talib

symbol = "APPL"
start_date = "2023-02-01"
end_date = "2023-03-29"

timeframe = "15m"

# Baixar dados do Yahoo Finance
data = yf.download(symbol, start=start_date, end=end_date, interval=timeframe)

# Calcular as Bandas de Bollinger
upper, middle, lower = talib.BBANDS(data['Close'], timeperiod=20, nbdevup=2, nbdevdn=2)

# Definir parâmetros de stop loss e take profit
stop_loss = 10
take_profit = 20

# Definir capital inicial e quantidade de ações em carteira
capital = 10000
acoes = 0

# Simular a estratégia
for i in range(len(data)):
    # Se o preço cruzou abaixo da banda inferior, comprar ações
    if data['Close'][i] < lower[i]:
        acoes_compradas = int(capital / data['Close'][i])
        acoes += acoes_compradas
        capital -= acoes_compradas * data['Close'][i]
    # Se o preço cruzou acima da banda superior, vender ações
    elif data['Close'][i] > upper[i]:
        capital += acoes * data['Close'][i]
        acoes = 0
    
    # Verificar se a posição atingiu o stop loss ou take profit
    if acoes > 0:
        j = i + 1
        while j < len(data) and (data['High'][j] < take_profit * data['Close'][i] and data['Low'][j] > (1 - stop_loss/100) * data['Close'][i]):
            j += 1
        if j < len(data) and data['High'][j] >= take_profit * data['Close'][i]:
            # A posição atingiu o take profit
            capital += acoes * take_profit * data['Close'][i]
            acoes = 0
        elif j < len(data) and data['Low'][j] <= (1 - stop_loss/100) * data['Close'][i]:
            # A posição atingiu o stop loss
            capital += acoes * stop_loss/100 * data['Close'][i]
            acoes = 0

# Calcular o capital final e rentabilidade
capital_final = capital + acoes * data['Close'][-1]
rentabilidade = (capital_final - capital) / capital * 100

print("Capital final: R$", round(capital_final, 2))
print("Rentabilidade:", round(rentabilidade, 2), "%")
```
